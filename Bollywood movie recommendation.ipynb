```python
import pandas as pd
```

```python
df = pd.read_csv("IMDB-Movie-Dataset(2023-1951) (1).csv")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df[['movie_id', 'movie_name', 'overview', 'genre', 'cast', 'director']]
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genre']+ df['cast'] + df['director']
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[['movie_id', 'movie_name', 'tags']]
```

```python
df.rename(columns = {'movie_name':'title'}, inplace = True)
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
